# **ASSIGNMENT ANSWERS**

**1. What is a parameter?**
A parameter is a numerical value that describes a whole population. It does not change unless the population changes. Example: average height of all students in a school.

**2. What is correlation?**
Correlation is a statistical relationship between two variables that shows how strongly they move together.

**3. What does negative correlation mean?**
Negative correlation means when one variable increases, the other decreases. They move in opposite directions.

**4. Define Machine Learning. What are the main components in Machine Learning?**
Machine Learning is a method where computers learn patterns from data without being directly programmed.
Main components are:

1. Data
2. Model
3. Loss function
4. Optimizer
5. Evaluation

**5. How does loss value help in determining whether the model is good or not?**
Loss value shows how far the model’s predictions are from the actual values.
Lower loss means the model is performing well, higher loss means the model is performing poorly.

**6. What are continuous and categorical variables?**
Continuous variables: numbers that can take any value (height, weight, temperature).
Categorical variables: data divided into groups or labels (gender, color, city).

**7. How do we handle categorical variables in Machine Learning? What are the common techniques?**
We convert categories into numbers.
Common techniques:

1. Label Encoding
2. One-Hot Encoding
3. Ordinal Encoding

**8. What do you mean by training and testing a dataset?**
Training dataset is used to teach the model.
Testing dataset is used to check how well the model performs on unseen data.

**9. What is sklearn.preprocessing?**
It is a module in Python that contains tools for scaling, encoding, normalizing, and transforming data.

**10. What is a Test set?**
A test set is a part of the dataset that is kept aside to evaluate the final model.

**11. How do we split data for model fitting (training and testing) in Python?**
We usually use the train_test_split function to divide the data into training data and testing data.

**12. How do you approach a Machine Learning problem?**

1. Understand the problem
2. Collect data
3. Clean and prepare data
4. Perform EDA
5. Split into training and testing
6. Select a model
7. Train the model
8. Evaluate the model
9. Improve with tuning

**13. Why do we have to perform EDA before fitting a model to the data?**
EDA helps us understand patterns, missing values, outliers, variable relationships, and data distribution.
It allows us to clean and prepare the data correctly before building the model.

**16. How can you find correlation between variables in Python?**
By using the `.corr()` function on a DataFrame.

**17. What is causation? Explain difference between correlation and causation with an example.**
Causation means one variable actually causes a change in another.
Correlation only shows a relationship, not cause.
Example: Ice cream sales and drowning cases increase together (correlation), but eating ice cream does not cause drowning (no causation).

**18. What is an Optimizer? What are different types of optimizers? Explain each with an example.**
Optimizer updates model parameters to reduce loss.
Types:

1. SGD – updates parameters using basic gradient direction.
2. Adam – faster optimizer that uses momentum and adaptive learning rate.
3. RMSProp – adjusts learning rate depending on recent gradients.

Example: Adam optimizer is commonly used while training neural networks.

**19. What is sklearn.linear_model?**
It is a module in sklearn that contains linear models like Linear Regression, Logistic Regression, etc.

**20. What does model.fit() do? What arguments must be given?**
model.fit() trains the model using data.
Arguments: features (X) and target (y).

**21. What does model.predict() do? What arguments must be given?**
model.predict() gives output for new input data.
Argument: new data (X).

**23. What is feature scaling? How does it help in Machine Learning?**
Feature scaling means transforming numerical values to a similar range.
It helps the model learn faster and improves accuracy because all features contribute equally.

**24. How do we perform scaling in Python?**
We use StandardScaler, MinMaxScaler, or RobustScaler from sklearn.preprocessing.

**25. What is sklearn.preprocessing?**
It is a module used for encoding, scaling, normalizing, and transforming data.

**26. How do we split data for model fitting (training and testing) in Python?**
By using train_test_split from sklearn.model_selection.

**27. Explain data encoding.**
Data encoding is the process of converting categorical (non-numeric) data into numeric form so that machine learning models can use it. Examples include Label Encoding and One-Hot Encoding.
